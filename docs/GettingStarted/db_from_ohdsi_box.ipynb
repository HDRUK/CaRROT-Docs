```python
import sqlalchemy as sql
import pandas as pd

ngin = sql.create_engine('postgresql://etldev:etldev@192.168.0.104:5432/postgres')
cnx = ngin.connect()

inspector = sql.inspect(ngin)

def get_df(schema,table_name):
    selection = r'''
SELECT
*
FROM %s.%s
'''%(schema,table_name)

    return pd.read_sql(selection,ngin)


for schema in inspector.get_schema_names():
    print (schema)

```

```python
for schema in inspector.get_schema_names():
    print (f'===== {schema} ===== ')
    for table in inspector.get_table_names(schema=schema):
        print ('\t ',table)
```
